```python
import pandas as pd

df = pd.read_csv('/Titanic-Dataset.csv')
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# Replace Null Values of Age column with Its Mean
mean_age = round(df['Age'].mean())
df['Age'].fillna(mean_age, inplace = True)

df
```

```python
# Replace Null Values of Cabin Column with unknown
df['Cabin'].fillna('unknown', inplace = True)
df
```

```python
# Encode using mapping
# Common values: S = 0, C = 1, Q = 2
embarked_mapping = {'S': 0, 'C': 1, 'Q': 2}

df['Embarked_code'] = df['Embarked'].map(embarked_mapping)

df
```

```python
# Min-Max Normalization
df['Age_normalized'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df['Fare_normalized'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())

# Z-Score Standardization
df['Age_standardized'] = (df['Age'] - df['Age'].mean()) / df['Age'].std()
df['Fare_standardized'] = (df['Fare'] - df['Fare'].mean()) / df['Fare'].std()

df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.boxplot(df['Age'])
plt.title('Box Plot for Fare')
plt.ylabel('Fare')
plt.show()
```

```python
#removing outliera
age_data = df['Age']

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = age_data.quantile(0.25)
Q3 = age_data.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
filtered_data=df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]

# Verify if outliers are removed
print(f'Original dataset size: {df.shape[0]}')
print(f'Filtered dataset size (outliers removed): {filtered_data.shape[0]}')
```
